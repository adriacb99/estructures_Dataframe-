```python
Dataframe["VelMitjana(km/h)"] = (Dataframe["Distance"])/(Dataframe["AirTime"]/60)
Dataframe['ArribadaTard'] = ["NO" if x == 0 else "SI" for x in Dataframe["LateAircraftDelay"]]
Dataframe['ArribadaTard'] = ["SI" if x > 0 else "NO" for x in Dataframe["LateAircraftDelay"]]
table = Dataframe[["FlightNum","TailNum", "Year", "Month", "DayofMonth", "DayOfWeek", "AirTime", "Distance", "VelMitjana(km/h)",'ArribadaTard']]
table
```
